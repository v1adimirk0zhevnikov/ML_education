**<p> Создайте массив из 20 случайных чисел. Найдите:</p>**
1) Минимальное и максимальное значения
2) Медиану
3) Стандартное отклонение

```python
def thirdSolution():
    arr: np.ndarray = np.random.random(20)
    print(f"Maximal value {np.max(arr)}")
    median = np.sum(np.sort(arr)[len(arr) // 2: len(arr) // 2 + 2]) / 2 
    print(f"Median: {median}")
    print(f"Standart deviation {np.std(arr)}")

thirdSolution()
```

**<p>Создайте матрицу произвольного размера.</p>**
1) Найдите транспонированную матрицу.
2) Найдите обратную матрицу, если это возможно.
3) Найдите собственные значения и собственные вектора.

```python
def fifthSolutionOne(rows: int = 5, cols: int = 5):
    matrix: np.ndarray = np.random.random((rows, cols)) * 100
    transposedMatrix = matrix.T
    print(f"Transposed matrix:\n{transposedMatrix}")
    try:
        inversed = np.linalg.inv(matrix)
        print(f"Inversed matrix:\n{inversed}")
    except np.linalg.LinAlgError:
        print("Can't inverse matrix")

    try:
        eigenvalues, eigenvectors = np.linalg.eig(matrix)
        print("Eigen values:")
        [print(value) for value in eigenvalues]
        print("Eigen vectors:")
        [print(vector) for vector in eigenvectors]
    except np.linalg.LinAlgError:
        print("Can't find eigenvalues and eigenvectors")


fifthSolutionOne(3, 3)
```

**<p>Создайте трёхмерный массив размером 3x3x3, заполненный<br>

**<p>Сгенерируйте набор данных из 100 случайных точек в 3D-<br>
пространстве. Выполните их нормализацию (приведите к<br>
единичному масштабу) по каждой оси.</p>**

```python
def seventhSolution():
    arr = np.random.random((100, 3)) * 100
    minValue = np.min(arr)
    maxValue = np.max(arr)
    normalizedArr = (arr - minValue) / (maxValue - minValue)
    print(f"Normalized array:\n{normalizedArr}")


seventhSolution()
```

**<p>Загрузите любое изображение (в рамках приличия, конечно) и<br>
сожмите его с помощью сингулярного разложения (SVD).<br>
Выберете разное количество сингулярных значений и сравните<br>
полученный результат с оригиналом.</p>**

```python
def eightSolution(path: str, k_values: list):
    # Загрузка изображения в BGR
    img_bgr = cv2.imread(path)
    if img_bgr is None:
        raise FileNotFoundError(f"Не удалось загрузить изображение по пути: {path}")
    
    # Преобразуем в RGB для корректного отображения в matplotlib
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    height, width, channels = img_rgb.shape

    # Отображаем оригинальное изображение
    plt.figure(figsize=(15, 10))
    plt.subplot(2, 3, 1)
    plt.imshow(img_rgb)
    plt.title("Оригинал")
    plt.axis("off")

    # Сжатие с разными k
    for idx, k in enumerate(k_values):
        # Создаём копию для сжатого изображения
        compressed = np.zeros_like(img_rgb, dtype=np.float64)

        # Применяем SVD к каждому каналу отдельно
        for ch in range(channels):
            U, S, Vt = np.linalg.svd(img_rgb[:, :, ch].astype(np.float64), full_matrices=False)
            # Восстанавливаем с использованием первых k сингулярных значений
            compressed[:, :, ch] = (U[:, :k] @ np.diag(S[:k]) @ Vt[:k, :])

        # Приводим к uint8 и ограничиваем [0, 255]
        compressed = np.clip(compressed, 0, 255).astype(np.uint8)

        # Отображаем
        plt.subplot(2, 3, idx + 2)
        plt.imshow(compressed)
        plt.title(f"k = {k}")
        plt.axis("off")

    plt.tight_layout()
    plt.show()

# Пример использования:
k_list = [5, 20, 50, 100, 200]
eightSolution("bebra.jpeg", k_list)
```
