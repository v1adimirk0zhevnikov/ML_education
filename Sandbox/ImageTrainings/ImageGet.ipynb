```python
import socket
import select
from PIL import Image
import io
import matplotlib.pyplot as plt
```

```python
def getImage(HOST, PORT):

    #Declare socket:
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    s.settimeout(5)

    try:
        s.connect((HOST, PORT))
        s.send(b'GET / HTTP/1.0 \n\n')
        data = s.recv(4096 * 2)

        rdy = select.select([s], [], [], 2)
        if not rdy[0]:
            raise RuntimeError("No response")
            return

        if not data:
            raise RuntimeError("Socket connection broken")
            
        return data
        
    except Exception as exc:
        print("Socket error:", str(exc))
        return

    finally:
        s.close()


def BytesToPNG(data):
    with open('output_1.png', 'wb') as image_file:
        image_file.write(data)

if __name__ == "__main__":
    HOST = "task.miminet.ru"
    PORT = 8010
    data = getImage(HOST, PORT)
    BytesToPNG(data)
```
