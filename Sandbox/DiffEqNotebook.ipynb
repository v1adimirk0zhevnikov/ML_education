```python
import numpy as np
import matplotlib.pyplot as plt
```

Изоклемы для y' = y - x^2

```python
def firstIsoclemm():
    n: int = 25
    m: int = 5
    constants: np.ndarray = np.linspace(-100, 100, n)
    x: np.ndarray = np.linspace(-10, 10, 400)
    y: np.ndarray = np.zeros(400 * n).reshape(n, 400)
    for i in range(len(constants)):
        y[i] = constants[i] + x ** 2
    
    fig, ax = plt.subplots()
    for i in range(len(constants)):
        ax.plot(x, y[i], color="red", linewidth=0.4)
        for j in range(m):
            idx: int = len(x) // m * j
            x_start = x[idx]
            y_start = y[i][idx]
            x_end = x[idx + 4]
            y_end = y[i][idx + 4]
            ax.annotate('', xy=(x_start, y_start), xytext=(x_end, y_end),
                        arrowprops=dict(facecolor='red', shrink=1, width=0.4, headwidth=2))
    
    ax.set_ylim(0, 100)
    ax.set_xlim(-10, 10)
    plt.show()


firstIsoclemm()
```

Изоклемы для y' = x - exp(y)

```python
def secondIsoclemm(constantsNumber: int = 50):
    
    constants: np.ndarray = np.linspace(-100, 100, constantsNumber)
    y: np.ndarray = np.linspace(-5, 5, 400)
    x: np.ndarray = np.zeros(constantsNumber * 400).reshape(constantsNumber, 400)

    fig, ax = plt.subplots()
    for i in range(constantsNumber):
        x[i] = constants[i] + np.exp(y)
        ax.plot(x[i], y, color="red")
        for j in range(m):
            idx: int = len(y) // m * j
            start: tuple = x[i][idx], y[idx]
            end: tuple = x[i][idx + 4], y[idx + 4]
            ax.annotate('', xy=start, xytext=end,
                        arrowprops=dict(facecolor='red', shrink=1, width=0.4, headwidth=2))
    
    ax.set_ylim(-5, 3.3)
    ax.set_xlim(-10, 10)
    ax.grid(linestyle="--", alpha=0.6, color="grey")
    plt.show()


secondIsoclemm()
```
