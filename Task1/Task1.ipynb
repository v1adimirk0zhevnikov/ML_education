```python
"""
Задание 1
-Реализуйте функцию bubble_sort(lst: list)
-*Посчитайте сложность алгоритма
"""

def bubble_sort(lst: list) -> list:
    length: int = len(lst)
    for i in range(length):
        for j in range(0, length - i - 1):
            if lst[j] > lst[j + 1]:
                lst[j], lst[j + 1] = lst[j + 1], lst[j]
    return lst


if __name__ == "__main__":
    l: list = [2, 3, 2, 1, 45, -12]
    print(bubble_sort(l))
```

```python
"""
Задание 2
-Реализуйте функцию binary_search
-*Посчитайте сложность алгоритма
"""
def binary_search(arr: list, x: int) -> int:
    """
    Бинарный поиск: возвращает индекс элемента x или -1, если его нет.
    """
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] > x:
            right = mid - 1
        elif arr[mid] < x:
            left = mid + 1
        elif arr[mid] == x:
            return mid
        else:
            break
    return -1

if __name__ == "__main__":
    array: list = list(range(1, 1000, 3))
    targets: list = (23, 232)

    print(binary_search(array, targets[0]))
    print(binary_search(array, targets[1]))
    #Сложность O(log(N))
```

```python
"""
Задание 3
-Реализуйте функцию pascal_triangle
-*Объяснить связь треугольника Паскаля с биномиальными коэффициентами
"""
def pascal_triangle(n: int) -> list:
    """
    Возвращает список списков с n строками треугольника Паскаля.
    """
    if n <= 0: 
        return []
    if n == 1:
        return [1]
    elif n == 2:
        return [[1], [1, 1]]
    triangle: list = [[1], [1, 1]]
    for row in range(3, n + 2):
        currentList: list = [1] * row
        for i in range(1, row - 1):
            currentList[i] = triangle[row-2][i-1] + triangle[row-2][i]
        triangle.append(currentList)
    return triangle


if __name__ == "__main__":
    n: int = 10
    triangle = pascal_triangle(8)
    for i in range(len(triangle)):
        print(*triangle[i])
# Связь с С: k коэффициент в n ряду выражается через n!/((n-k)! * k!)
```

```python
"""
Задание 4
-Написать любой способ
-*Написать два способа
"""
def factorial_recursive(n: int) -> int:
    """
    Вычисляет факториал числа n рекурсивно.
    """
    # базовый случай: 0! = 1
    if (n == 1) or (n == 0):
        return 1
    return n * factorial_recursive(n - 1)
    
    
def factorial_iterative(n: int) -> int:
    """
    Вычисляет факториал числа n через цикл.
    """
    if (n == 0): return 1
    result: int = 1
    for i in range(1, n + 1):
        result *= i
    return result


if __name__ == "__main__":
    print(factorial_recursive(9))
    print(factorial_iterative(500))
#Два способа
```

```python
"""
Задание 5
-Реализовать функцию is_prime
-*Объяснить, почему проверяем числа до sqrt(n) (квадратного корня от n)
"""
def is_prime(n: int) -> bool:
    """
    Проверяет, простое ли число n.
    """
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 2):
        if not(n % i) : return False
    return True

if __name__ == "__main__":
    print(is_prime(11))
#Проверяем до кв корня т.к. числа большие кв корня не могут быть сомножителями
```

```python
"""
Задание 6
-Реализовать функцию knapsack
"""
def knapsack(W: int, weights: list, values: list, n: int) -> int:
    """
    Решение задачи о рюкзаке методом динамического программирования.
    """
    DP: list[list] = [ [0] * (W + 1) for _ in range(n + 1) ]
    # создать таблицу DP размера (n+1) x (W+1)
    # заполнить её по правилам выбора предметов
    for i in range(1, n + 1):
        for w in range(W + 1):
            if (w == 0):
                DP[i][w] = 0
            elif weights[i-1] <= w:
                DP[i][w] = max(DP[i-1][w], values[i-1] + DP[i-1][w - weights[i-1]])
            else:
                DP[i][w] = DP[i-1][w]
    print(DP)
    return DP[n][W]


if __name__ == "__main__":
    print(knapsack(5, [1, 2, 3, 4], [3, 2, 1, 3], 4))
```
